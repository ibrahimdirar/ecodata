```python
import datetime
import pandas as pd
import numpy as np
import sqlite3
import json

from dataclasses import dataclass

@dataclass
class EcoData:
    ticker: str
    date: str
    original: float
    revised: float

db = sqlite3.connect('data.db')

db.execute('CREATE TABLE IF NOT EXISTS eco_data_internal (ticker, date, original, revised)')
db.execute('CREATE TABLE IF NOT EXISTS eco_data_external (ticker, date, original, revised)')

# for last date in month from 2010-01-01 to 2022-9-30
# create a list of EcoData objects
eco_datas = []
for date in pd.date_range(start='2010-01-01', end='2022-9-30', freq='M'):
    # create EcoData object
    # number of months until 2022-9-30
    months_to_end = (pd.to_datetime('2022-9-30') - date).days // 30
    revised = None if months_to_end <= 1 else np.random.randint(100, 200)
    # get last working day of month
    eco_data = EcoData(
        ticker='GB',
        date=date.strftime('%Y-%m-%d'),
        original=np.random.randint(100, 200),
        revised=revised,
        )
    print(eco_data)
    # save object to database
    if months_to_end >= 1:
        db.execute('INSERT INTO eco_data_internal VALUES (:ticker, :date, :original, :revised)', eco_data.__dict__)
    # adjust date to last week day of the month
    eco_data.date = (date - pd.offsets.Day(5) + pd.offsets.BMonthEnd(0)).strftime('%Y-%m-%d')
    if 1 <= months_to_end <= 3:
        eco_data.revised = np.random.randint(100, 200)
    db.execute('INSERT INTO eco_data_external VALUES (:ticker, :date, :original, :revised)', eco_data.__dict__)

db.commit()
```
